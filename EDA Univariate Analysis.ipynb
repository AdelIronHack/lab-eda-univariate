```python
#1. Generate a frequency table for the product `category`
frequency_table = df['category'].value_counts()
frequency_table
```

```python
# Which are the top 5 most listed product categories?
frequency_table = frequency_table.head(5)
frequency_table
```

```python
# 2. Display the distribution of products across different categories using a bar chart.*If you face problems understanding the chart, do it for a 
# subset of top categories
sns.barplot(x=frequency_table.values, y=frequency_table.index, palette="Set3")
```

```python
# For a subset of top categories, visualize their proportions using a pie chart. Does any category dominate the listings?
frequency_table.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
```
